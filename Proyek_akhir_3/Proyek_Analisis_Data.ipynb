## Visualization & Explanatory Analysis

### Pertanyaan 1: In what season are the most and least bicycles rented?

```python
bike_day_df.replace({
    "season": {1: "Springer", 2: "Summer", 3: "Fall", 4: "Winter"},
    "yr": {0: 2011, 1: 2012},
    "holiday": {0: "Not Holiday", 1: "Holiday"},
    "workingday": {0: "Holiday", 1: "Working Day"},
    "weekday": {0: "Sunday", 1: "Monday", 2: "Tuesday", 3: "Wednesday", 4: "Thursday", 5: "Friday", 6: "Saturday"}
    }, inplace=True)

plot_season = bike_day_df.groupby('season', observed=False)[['registered', 'casual']].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(
    data=plot_season, 
    x='season', 
    y='registered', 
    label='Registered', 
    color='tab:blue'
)
sns.barplot(
    data=plot_season, 
    x='season', 
    y='casual', 
    label='Casual', 
    color='tab:green'
)
plt.title('Total Bikes Rented in the Last 2 Years Based on Season')
plt.xlabel(None)
plt.ylabel(None)
plt.legend()
plt.show()
```

The following is a mappin number to season name:
1. jumper
2. summer
3. fall
4. winter

The highest total bicycle rentals occur in the fall season for both casual and registered customers. On the other hand, the total number of bicycle rentals is the lowest in Musing Springer.

### Pertanyaan 2: How does the season affect the total number of bicycle rentals?

```python
bike_day_df.info()
```

```python
season_counts = bike_day_df.groupby(by=["season", "year"]).agg({
    "total_sum": "sum"
}).reset_index()  

sns.barplot(data=season_counts, x="season", y="total_sum", hue="year", palette=["blue", "green"])
plt.ylabel("Jumlah")
plt.title("Jumlah total sepeda yang disewakan berdasarkan musim")
plt.legend(title="Tahun", loc="upper right")  
plt.tight_layout()
plt.show()
```

The highest total bicycle rentals occur in the fall season for both casual and registered customers. On the other hand, the total number of bicycle rentals is the lowest in Musing Springer.

### Pertanyaan 3: How is the performance of rental bicycle lending?

```python
bike_day_df.rename(columns={
    'dteday' : 'date',
    'yr' : 'year',
    'mnth' : 'month',
    'weathersit' : 'weather',
    'cnt' : 'total'
}, inplace=True)

mapping_month = {
    1: 'Jan', 2: 'Feb', 3: 'Mar',
    4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep',
    10: 'Oct', 11: 'Nov', 12: 'Dec'   
}
bike_day_df['month'] = bike_day_df['month'].map(mapping_month)
```

```python


plot_month = bike_day_df['month'].astype(str)
plot_year = bike_day_df['year'].astype(str)

bike_day_df['year_month'] = plot_month + ' ' + plot_year

bike_day_df['total_sum'] = bike_day_df.groupby('year_month')['total'].transform('sum')

plt.figure(figsize=(20, 6))
sns.set(style='whitegrid')
sns.lineplot(
    data=bike_day_df, 
    x='year_month', 
    y='total_sum',  
    marker='o'
)
plt.title("Total Bikes Rented monthly")
plt.xlabel(None)
plt.ylabel(None)
plt.xticks(rotation=45)
plt.show()
```

There was a fairly high increase in interest in renting bicycles in 2012 when compared to the previous year.
